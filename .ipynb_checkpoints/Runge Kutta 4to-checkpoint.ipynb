```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
```

```python
t,y,v=sp.symbols('t,y,v')
f=-0.9*y*(v/v)
y0=0
v0=20
tf=100
dt=0.01
```

```python
f1=sp.lambdify([y,v],f)
ts,ys,vs,ac=RK4(y0,v0,dt,tf)
plt.plot(ts,ys,color='black',label='posicin')
plt.plot(ts,vs,color='red',label='Velocidad')
plt.plot(ts,ac,color='blue',label='Aceleracin')
plt.grid('on')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
def dydt(y):
    return np.array([y[1],f1(y[0],y[1])])
```

```python
def RK4(y0,v0,dt,tf):
    ts=np.arange(0,tf,dt)
    n=len(ts)
    ys=ts*0
    vs=ts*0
    ac=ts*0
    ys[0]=y0
    vs[0]=v0
    for i in range(0,n-1):
        z0=np.array([ys[i],vs[i]])
        k1=dydt(z0)
        k2=dydt(z0+(dt*k1)/2)
        k3=dydt(z0+(dt*k2)/2)
        k4=dydt(z0+dt*k3)
        
        ys[i+1]=ys[i]+(dt/6)*(k1[0]+2*k2[0]+2*k3[0]+k4[0])
        vs[i+1]=vs[i]+(dt/6)*(k1[1]+2*k2[1]+2*k3[1]+k4[1])
        ac[i]=f1(ys[i],vs[i])
    return ts,ys,vs,ac
```
