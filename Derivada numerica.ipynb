```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
```

```python
      
def df(y,dt,ti,tf):
    t=np.arange(ti,tf,dt)
    n=len(y)
    v1=np.ones(n-1)*(1/2)
    v2=np.ones(n-1)*(-1/2)
    D1=np.diag(v1,k=1)
    D2=np.diag(v2,k=-1)
    D=(D1+D2)/dt
    yp=np.matmul(D, y)
    yp[0]=yp[1]
    yp[-1]=yp[-2]
    ##Plot##############################################################
    plt.plot(t,y,color='black',label='funcion')
    plt.plot(t,yp,color='red',label='Derivada numerica')
    #plt.axis('equal')
    plt.grid('on')
    plt.legend()
    plt.show()
```

```python
xs=np.linspace(-np.pi,4*np.pi,500)

y=np.sin(xs)
df(y,np.diff(xs)[0],min(xs),max(xs))
```

```python
def df2(y,dt):
    N=len(y)
    yp=y*0
    for i in range(1,N-1):
        yp[i]=(y[i+1]-2*y[i]+y[i-1])/dt**2
        
    yp[0]=yp[1]
    yp[-1]=yp[-2]
    return yp
```

```python
dx=0.0001
xs=np.arange(-2*np.pi,2*np.pi,dx)
y=np.exp(-xs**2)
plt.plot(xs,df2(y,dx),color='red')
x=sp.symbols('x')
f=sp.lambdify(x,sp.diff(sp.diff(sp.exp(-x**2),x),x))
plt.plot(xs,f(xs))
```
