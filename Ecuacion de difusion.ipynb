```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
k=1

tf=2 #timepo final
L=1 # Longitud de la varilla
nx=30 #numero de puntos en la varilla
dx=L/(nx)
dt=dx**2/4
ts=np.arange(0,tf,dt)

nt=len(ts)

T=np.zeros((nx, nt))
#condiciones iniciales
xs=np.arange(0,L,dx)
TT,X=np.meshgrid(ts,xs)
T[:,0]=np.sin(np.pi*xs)

for j in range(0,nt-1):
    for i in range(1,nx-1):
        T[i,j+1]=(dt/dx**2)*T[i-1,j]+(1-2*(dt/dx**2))*T[i,j]+(dt/dx**2)*T[i+1,j]
        #T[i,j+1]=((k*2*dt)/(dx**2))*(T[i-1,j]-2*T[i,j]+T[i+1,j])+T[i,j-1]
```

```python
SA=plt.contourf(np.sin(np.pi*X)*np.exp(-(np.pi**2)*TT))
plt.colorbar(SA)
plt.xlabel('Tiempo')
plt.ylabel('Espacio')
plt.title('Solucion analitica')
```
