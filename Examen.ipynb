```python
def S(u,t):
    dt=np.diff(t)[0]
    f=abs(u)**2-2*abs(u)**4+2*abs(df(u,dt))**2
    return trapz(t,f)
```

```python
def df(y,dt):
    yp=y*0
    n=len(y)
    for i in range(1,n-1):
        yp[i]=(y[i+1]+y[i-1])/(2*dt)
    yp[0]=yp[1]
    yp[-1]=yp[-2]
    return yp
```
