```python
import numpy as np
import sympy as sp
from sympy import nroots
from matplotlib import pyplot as plt
```

```python
x,y,z=sp.symbols('x,y,z')
```

```python
r=sp.symbols('r')
#r=sp.sqrt(x**2+y**2)
k=7
l0=2
showc=1
```

```python
F=-k*(r-l0)+(k/5)*(r-l0)**3
roots=nroots(F)
```

```python
r=sp.sqrt(x**2+y**2)
F=-k*(r-l0)+(k/5)*(r-l0)**3
Fgay=sp.lambdify([x,y],F)####################################magnitud de la fuerza
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
Fx=F*(x/r)
Fy=F*(y/r)
xx=np.linspace(-float(max(roots))-1,float(max(roots))+1,100)
X,Y=np.meshgrid(xx,xx)

fx=sp.lambdify([x,y],Fx)
fy=sp.lambdify([x,y],Fy)

FXX=fx(X,Y)
FYY=fy(X,Y)
FF=np.sqrt(FXX**2+FYY**2)


ax.quiver(X,Y,FXX/FF,FYY/FF)
if showc==1:
    circle = plt.Circle((0,0), roots[0], color='b', fill=False)
    circle2 = plt.Circle((0,0), roots[1], color='b', fill=False)
    circle3=plt.Circle((0,0), roots[2], color='b', fill=False)
    ax = plt.gca()
    ax.add_patch(circle)
    ax.add_patch(circle2)
    ax.add_patch(circle3)

#plt.streamplot(X,Y,FXX/FF,FYY/FF)
#plt.axis('equal')
plt.grid('on')
```

```python
Fgay(X,Y)
cp=plt.contourf(Fgay(X,Y),levels=300,cmap='jet')
plt.colorbar(cp)
```
