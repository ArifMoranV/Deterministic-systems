```python
import numpy as np 
import sympy as sp
from matplotlib import pyplot as plt
```

```python
li=-3
ls=3
x=np.linspace(li,ls,100)
[X,Y]=np.meshgrid(x,x)
```

```python
#vestores 
dxdt=Y
dydt=2*X*(1-X**2)


fig, ax = plt.subplots(figsize = (10, 10))
n=np.sqrt(dxdt**2+dydt**2)
#ax.quiver(X,Y,dxdt/n,dydt/n)
ax.streamplot(X,Y,dxdt/n,dydt/n,color='red')
plt.xlim([li+1,ls-1])
plt.ylim([li+1,ls-1])
plt.legend()
plt.axis('equal')
ax.grid('on')
plt.show()
```
