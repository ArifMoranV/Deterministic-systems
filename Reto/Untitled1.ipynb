```python
import numpy as np
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt

# Example discrete oscillating function (replace with your own data)
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Find local maxima and minima indices
local_maxima_indices = argrelextrema(y, np.greater)[0]
local_minima_indices = argrelextrema(y, np.less)[0]

# Plot original function with peaks
plt.plot(x, y, label='Function')
plt.plot(x[local_maxima_indices], y[local_maxima_indices], 'ro', label='Peaks')
plt.plot(x[local_minima_indices], y[local_minima_indices], 'bo', label='Troughs')
plt.legend()
plt.show()
```
