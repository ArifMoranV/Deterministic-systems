```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt

```

```python
x=sp.symbols('x')
f=sp.sin(20*x)+sp.cos(20*x)+sp.sin(55*x)+sp.cos(500*x)+sp.sin(500*x)
f1=sp.lambdify(x,f)
t=np.linspace(0,2*np.pi,1000)
ys=f1(t)
DFT(ys)
```

```python
def DFT(y):
    N=len(y)
    M=np.zeros((N,N))
    for n in range(N):
        for k in range(N):
            M[k,n]=np.real(np.exp((-1j*2*np.pi*k*n)/N))
            
    ns=np.arange(0,N)
    yh=(np.matmul(M,y))
    plt.plot(ns,yh,label='Transformada discreta de fourier')
    plt.legend()
    plt.grid('on')
```
